## Theory: What Makes a Great Capstone Project?

### The Purpose of Capstone Projects

Capstone projects serve several critical purposes:

1. **Integration**: Combining multiple ML techniques and concepts learned throughout the course
2. **Demonstration**: Showcasing your ability to handle end-to-end ML workflows
3. **Portfolio Building**: Creating tangible evidence of your skills for potential employers or collaborators
4. **Problem-Solving**: Applying ML to solve real-world problems with practical impact

### Project Planning Framework

A successful capstone project follows a structured approach:

#### 1. Problem Formulation

The first step is clearly defining your problem. In machine learning, we typically formulate problems as:

- **Classification**: Predicting categorical labels
- **Regression**: Predicting continuous values
- **Clustering**: Grouping similar data points
- **Recommendation**: Suggesting items based on preferences

#### 2. Success Metrics Selection

Choosing appropriate evaluation metrics is crucial. The metrics you select depend on your problem type and business objectives.

For **Classification Problems**:

$$\text{Accuracy} = \frac{\text{Correct Predictions}}{\text{Total Predictions}} = \frac{TP + TN}{TP + TN + FP + FN}$$

$$\text{Precision} = \frac{TP}{TP + FP}$$

$$\text{Recall} = \frac{TP}{TP + FN}$$

$$F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$$

For **Regression Problems**:

$$\text{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

$$\text{RMSE} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$$

$$\text{MAE} = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$$

$$R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

#### 3. Dataset Selection Criteria

A good capstone dataset should:

- Be **sufficiently complex** (multiple features, interesting patterns)
- Have **adequate size** (enough samples for training and testing)
- Be **relevant** to your interests or career goals
- Present **real challenges** (missing data, imbalanced classes, etc.)
- Have **clear documentation** and provenance

### Project Planning Checklist

Let's create a structured checklist to guide your project planning:

```python
class CapstoneProjectPlanner:
    """
    A helper class to plan and track capstone project components.
    """
    
    def __init__(self, project_name):
        self.project_name = project_name
        self.components = {
            'Problem Definition': False,
            'Dataset Selection': False,
            'Success Metrics Defined': False,
            'Exploratory Data Analysis': False,
            'Feature Engineering': False,
            'Baseline Model': False,
            'Advanced Model': False,
            'Model Evaluation': False,
            'Documentation': False,
            'Presentation': False
        }
    
    def complete_component(self, component_name):
        """Mark a component as complete."""
        if component_name in self.components:
            self.components[component_name] = True
            print(f"✓ {component_name} completed!")
        else:
            print(f"Component '{component_name}' not found.")
    
    def get_progress(self):
        """Calculate project completion percentage."""
        completed = sum(self.components.values())
        total = len(self.components)
        return (completed / total) * 100
    
    def show_status(self):
        """Display current project status."""
        print(f"\n{'='*50}")
        print(f"Project: {self.project_name}")
        print(f"Progress: {self.get_progress():.1f}%")
        print(f"{'='*50}\n")
        
        for component, status in self.components.items():
            status_symbol = "✓" if status else "○"
            print(f"{status_symbol} {component}")

# Example usage
my_project = CapstoneProjectPlanner("Customer Churn Prediction")
my_project.show_status()

# Simulate completing some components
print("\nCompleting initial phases...\n")
my_project.complete_component('Problem Definition')
my_project.complete_component('Dataset Selection')
my_project.complete_component('Success Metrics Defined')

print(f"\nCurrent progress: {my_project.get_progress():.1f}%")
```

### Evaluation Metrics Calculator

Let's build a tool to help you understand and calculate different evaluation metrics:

```python
class MetricsCalculator:
    """
    Calculate and display various ML evaluation metrics.
    """
    
    @staticmethod
    def classification_metrics(y_true, y_pred, class_names=None):
        """
        Calculate classification metrics.
        
        Parameters:
        -----------
        y_true : array-like
            True labels
        y_pred : array-like
            Predicted labels
        class_names : list, optional
            Names of classes for the report
        """
        from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
        
        accuracy = accuracy_score(y_true, y_pred)
        precision = precision_score(y_true, y_pred, average='weighted')
        recall = recall_score(y_true, y_pred, average='weighted')
        f1 = f1_score(y_true, y_pred, average='weighted')
        
        print("Classification Metrics:")
        print(f"Accuracy:  {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall:    {recall:.4f}")
        print(f"F1-Score:  {f1:.4f}")
        
        return {'accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1': f1}
    
    @staticmethod
    def regression_metrics(y_true, y_pred):
        """
        Calculate regression metrics.
        
        Parameters:
        -----------
        y_true : array-like
            True values
        y_pred : array-like
            Predicted values
        """
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        
        print("Regression Metrics:")
        print(f"MSE:  {mse:.4f}")
        print(f"RMSE: {rmse:.4f}")
        print(f"MAE:  {mae:.4f}")
        print(f"R²:   {r2:.4f}")
        
        return {'mse': mse, 'rmse': rmse, 'mae': mae, 'r2': r2}

# Example: Classification metrics with sample data
print("Example 1: Classification Problem\n")
y_true_class = np.array([0, 1, 1, 0, 1, 1, 0, 0, 1, 0])
y_pred_class = np.array([0, 1, 1, 0, 0, 1, 0, 1, 1, 0])

calc = MetricsCalculator()
class_metrics = calc.classification_metrics(y_true_class, y_pred_class)

print("\n" + "="*50 + "\n")

# Example: Regression metrics with sample data
print("Example 2: Regression Problem\n")
y_true_reg = np.array([3.5, 2.8, 4.2, 5.1, 3.9, 4.7, 2.3, 5.5, 4.0, 3.2])
y_pred_reg = np.array([3.3, 2.9, 4.5, 5.0, 3.7, 4.8, 2.5, 5.3, 4.1, 3.4])

reg_metrics = calc.regression_metrics(y_true_reg, y_pred_reg)
```

### Dataset Quality Assessment Tool

Before committing to a dataset, assess its suitability for your capstone project:

```python
def assess_dataset_quality(df, target_column=None):
    """
    Assess the quality and suitability of a dataset for a capstone project.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The dataset to assess
    target_column : str, optional
        Name of the target variable column
    """
    print("Dataset Quality Assessment")
    print("="*50)
    
    # Basic information
    n_rows, n_cols = df.shape
    print(f"\n1. Size: {n_rows:,} rows × {n_cols} columns")
    
    # Check for missing values
    missing_counts = df.isnull().sum()
    missing_pct = (missing_counts / len(df)) * 100
    has_missing = missing_counts.sum() > 0
    
    print(f"\n2. Missing Values: {'Yes' if has_missing else 'No'}")
    if has_missing:
        missing_cols = missing_pct[missing_pct > 0].sort_values(ascending=False)
        print("   Columns with missing data:")
        for col, pct in missing_cols.head(5).items():
            print(f"   - {col}: {pct:.2f}%")
    
    # Data types
    print(f"\n3. Data Types:")
    dtype_counts = df.dtypes.value_counts()
    for dtype, count in dtype_counts.items():
        print(f"   - {dtype}: {count} columns")
    
    # Class balance (if target is specified)
    if target_column and target_column in df.columns:
        print(f"\n4. Target Variable Distribution ({target_column}):")
        value_counts = df[target_column].value_counts()
        for value, count in value_counts.items():
            pct = (count / len(df)) * 100
            print(f"   - {value}: {count} ({pct:.2f}%)")
    
    # Overall assessment
    print(f"\n5. Overall Assessment:")
    
    score = 0
    criteria = []
    
    if n_rows >= 1000:
        score += 1
        criteria.append("✓ Sufficient sample size")
    else:
        criteria.append("○ Small sample size (consider getting more data)")
    
    if n_cols >= 5:
        score += 1
        criteria.append("✓ Multiple features available")
    else:
        criteria.append("○ Limited features")
    
    if missing_pct.max() < 50:
        score += 1
        criteria.append("✓ Manageable missing data")
    else:
        criteria.append("○ High percentage of missing data")
    
    for criterion in criteria:
        print(f"   {criterion}")
    
    print(f"\n   Suitability Score: {score}/3")
    
    if score >= 2:
        print("   Recommendation: Good candidate for capstone project!")
    else:
        print("   Recommendation: Consider finding a more robust dataset.")

# Create a sample dataset for demonstration
np.random.seed(42)
sample_data = pd.DataFrame({
    'age': np.random.randint(18, 70, 1500),
    'income': np.random.randint(20000, 150000, 1500),
    'credit_score': np.random.randint(300, 850, 1500),
    'years_employed': np.random.randint(0, 40, 1500),
    'num_accounts': np.random.randint(1, 10, 1500),
    'loan_approved': np.random.choice([0, 1], 1500, p=[0.35, 0.65])
})

# Introduce some missing values
sample_data.loc[np.random.choice(sample_data.index, 50), 'credit_score'] = np.nan
sample_data.loc[np.random.choice(sample_data.index, 30), 'income'] = np.nan

# Assess the dataset
assess_dataset_quality(sample_data, target_column='loan_approved')
```

### Confusion Matrix Visualization

Understanding confusion matrices is crucial for evaluating classification models:

```python
# Create sample predictions for demonstration
np.random.seed(42)
y_true = np.random.randint(0, 3, 100)  # 3 classes
y_pred = y_true.copy()
# Introduce some errors
error_indices = np.random.choice(100, 20, replace=False)
y_pred[error_indices] = np.random.randint(0, 3, 20)

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Visualize
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True,
            xticklabels=['Class 0', 'Class 1', 'Class 2'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.title('Confusion Matrix Example', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.show()

print("\nInterpretation:")
print("- Diagonal elements represent correct predictions")
print("- Off-diagonal elements represent misclassifications")
print("- Larger diagonal values indicate better model performance")
```

## Key Takeaways

- **Capstone projects integrate all aspects of the ML workflow** from problem formulation to model evaluation and presentation.

- **Proper planning is essential**: Define your problem clearly, select appropriate metrics, and choose a dataset that provides sufficient complexity without being overwhelming.

- **Success metrics should align with your problem type**: Use classification metrics (accuracy, precision, recall, F1) for classification problems and regression metrics (MSE, RMSE, MAE, R²) for regression problems.

- **Dataset quality matters**: Ensure your dataset has sufficient size, manageable missing data, and presents real-world challenges that will showcase your skills.

- **Track your progress systematically**: Use planning tools and checklists to ensure you complete all necessary components of a comprehensive capstone project.
